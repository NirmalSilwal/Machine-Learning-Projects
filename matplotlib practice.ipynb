```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
x
```

```python
y=x**2
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([.1,.1,.8,.8])
```

```python
fig = plt.figure()

axes = fig.add_axes([.1,.1,.8,.8])

axes.plot(x,y)
```

```python
fig = plt.figure()

axes = fig.add_axes([.1,.1,.8,.8])

axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig=plt.figure()
     
axes1 = fig.add_axes([.1,.11,.8,.80])
axes2 = fig.add_axes([.21,.51,.4,.31])
# axes3 = fig.add_axes([.4,.2,.3,.2])
# left, bottom, width, height
axes1.plot(x,y)
axes1.set_title('larger plot')
axes2.plot(y,x)
axes2.set_title('smaller plot')
```

```python
fig=plt.figure(figsize=(5,7))
# (width,height)
axes1=fig.add_axes([.2,.3,.4,.5])
axes1.plot(x,y,'r')
```
