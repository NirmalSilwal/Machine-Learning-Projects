```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df=pd.DataFrame({'sin': np.arange(0, 5*np.pi, 0.01), 
                   'cos': np.arange(0.5*np.pi, 5.5*np.pi, 0.01)})
```

```python
df.columns
```

```python
df=np.tan(df)
```

```python
df.plot()
```

```python
def iqr(col):
    q1 = col.quantile(.25)
    q3 = col.quantile(.75)
    return q3 - q1 

# df.apply() executes the given function on a whole row or column
df.apply(iqr)
```

```python
dff=pd.DataFrame(np.ones([40,2]),dtype='int')
```

```python
def hello(val):
    q1 = val.quantile(.25)
    q3 = val.quantile(.75)
    return q3-q1
dff.apply(hello).plot()
```

```python
def add(val):
    return val+2

dff.applymap(add).plot()
```
